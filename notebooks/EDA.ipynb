```python
def drop_col_with_one_val(df):
    for column in df.columns:
        uniqs = df[column].unique()
        if len(uniqs) == 1:
            df = df.drop(column, axis=1)
    return df
```

```python
train_df = drop_col_with_one_val(train_df)
```
